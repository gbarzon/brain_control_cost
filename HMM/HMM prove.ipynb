```python
import numpy as np
import matplotlib.pyplot as plt

import ssm
```

```python
ts = np.arange(100)

data = np.sin(ts/np.pi) + np.sin(5 * ts/np.pi) + np.sin(0.1 * ts/np.pi) + 10 * np.random.random(len(ts))
data = np.expand_dims(data, axis=0)
```

```python
plt.plot(data[0])
plt.show()
```

```python
data.shape
```

```python
model = ssm.HMM(K=2, D=data.shape[0], observations="gaussian")
```

```python
hmm_lls = model.fit(datas=data.T, method="em")
```

```python
plt.plot(hmm_lls)
#plt.yscale('log')
plt.show()
```

```python
most_likely_states = model.most_likely_states(data.T)
most_likely_states
```

```python
plt.plot(data[0])
plt.plot(most_likely_states)

plt.show()
```
